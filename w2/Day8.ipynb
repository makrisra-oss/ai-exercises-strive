```
import requests
import pandas as pd
from bs4 import BeautifulSoup


page = requests.get("https://weather.com/weather/tenday/l/San+Francisco+CA?canonicalCityId=dfdaba8cbe3a4d12a8796e1f7b1ccc7174b4b0a2d5ddb1c8566ae9f154fa638c")
forecast = BeautifulSoup(page.content, 'html.parser')

temperatures = forecast.find_all('span', class_='DailyContent--temp--_8DL5')
temperature_list = []
temperature_high = []
temperature_low = []
temperature = temperature_high
for element in temperatures:
    variable = (int(element.get_text().strip('Â°')) - 32) * 5 / 9
    temperature_list.append(round(variable, 0))
for element in temperature_list:
    if temperature_list.index(element) % 2 == 0:
        temperature_high.append(element)
    else:
        temperature_low.append(element)

conditions = forecast.find_all('p', class_='DailyContent--narrative--3AcXd')
conditions_list = []
conditions_during_day = []
conditions_during_night = []
for element in conditions:
    conditions_list.append(element.get_text())
for element in conditions_list:
    if conditions_list.index(element) % 2 == 0:
        conditions_during_day.append(element)
    else:
        conditions_during_night.append(element)


#soup =BeautifulSoup(page.content, 'html.parser')
#print(soup.prettify())

# outer = soup.find_all('p', class_='outer-text')
#outer = soup.find(id='first').string
#print(outer)

week_days = forecast.find_all('span', class_='DailyContent--daypartDate--3MM0J')
week_day = []
for element in week_days:
    week_day.append(element.get_text())



data = {'temperature high': temperature_high,
'temperature low': temperature_low,
'conditions during the day': conditions_during_day,
'conditions during the night': conditions_during_night}
labels = week_day[::2]
df = pd.DataFrame(data, index= labels)

df.head(10)

```
